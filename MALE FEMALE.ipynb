```python

import pandas as pd
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
df = pd.read_csv("male female.csv")
```

```python
print("Dataset preview:\n", df.head())
```

```python
X = df[['height', 'weight', 'shoe size']]
y = df['gender']
```

```python
model = GaussianNB()
model.fit(X, y)
```

```python
probabilities = model.predict_proba(X)
```

```python
print("\nIndividual probabilities:")
for i, probs in enumerate(probabilities):
    print(f"Row {i+1}: Male = {probs[model.classes_.tolist().index('male')]:.2f}, Female = {probs[model.classes_.tolist().index('female')]:.2f}")
```

```python
avg_probs = probabilities.mean(axis=0)
print(f"\nAverage Probability:\nMale: {avg_probs[model.classes_.tolist().index('male')]:.2f}\nFemale: {avg_probs[model.classes_.tolist().index('female')]:.2f}")
```
